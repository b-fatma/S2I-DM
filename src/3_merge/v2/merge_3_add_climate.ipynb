```python
fire_landcover_elevation_soil_path = '/content/drive/MyDrive/dm_fire_prediction/merge/fire_landcover_elevation_no_binning_soil_no_na.csv'
climate_path = '/content/drive/MyDrive/dm_fire_prediction/premerge/climate_seasonal.csv'
```

```python
df = pd.read_csv(fire_landcover_elevation_soil_path)
gdf = gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(df.longitude, df.latitude))
gdf.head()
```

```python
gdf.describe()
```

```python
climate_df = pd.read_csv(climate_path)
climate_gdf = gpd.GeoDataFrame(climate_df, geometry=gpd.points_from_xy(climate_df.longitude, climate_df.latitude))
climate_gdf.head()
```

**Convert to integers**

```python
cols_to_convert = [*[x for x in climate_df.columns if x.startswith('tmax')], *[x for x in climate_df.columns if x.startswith('tmin')], *[x for x in climate_df.columns if x.startswith('prec')]]
climate_gdf[cols_to_convert] = climate_gdf[cols_to_convert].astype(int)
climate_gdf.head()
```

```python
climate_df.describe()
```

```python
joined = gdf.sjoin_nearest(
    climate_gdf,
    how='left'
)

joined
```

```python
joined.info()
```

```python
joined.drop(columns=['geometry', 'index_right', 'longitude_right', 'latitude_right'], inplace=True)
joined.rename({'longitude_left': 'longitude',
              'latitude_left': 'latitude'}, axis=1, inplace=True)
joined.head()
```
